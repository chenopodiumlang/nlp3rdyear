```python
import csv
```

```python
golden_standard_list = []
```

```python
with open('golden_standard.csv', encoding='utf-8') as stand_file:
    stand_reader = csv.reader(stand_file, delimiter=',')
    for row in stand_reader:
        word = [row[0], row[1]]
        golden_standard_list.append(word)
print(golden_standard_list)
        
```

```python
def convert_to_standard(word, parser):
    convert_dict = {
        ('ADJF', 'ADJS', 'COMP', 'A', 'ADJ'): 'ADJ',
        ('ADV', 'ADVB', 'ADVPRO'): 'ADV',
        ('NUM', 'NUMR', 'ANUM'): 'NUM',
        ('PRON', 'NPRO', 'APRO', 'SPRO'): 'PRON',
        ('S', 'NOUN', 'PROPN'): 'NOUN',
        ('V', 'VERB', 'INFN', 'AUX', 'GRND', 'PRED'): 'VERB',
        ('PRTF', 'PRTS'): 'PRT',
        ('CONJ', 'SCONJ', 'CCONJ'): 'CONJ',
        ('INTJ'): 'INTJ',
        ('PART, PRCL'): 'PART',
        ('PR', 'PREP', 'ADP'): 'PREP'
    }
    if parser == 'pymorphy':
        pos = str(word.tag).split(',')[0]
        for key in convert_dict:
            if pos in key:
                pos = convert_dict[key]
        
    if parser == 'spacy':
        pos = word
        for key in convert_dict:
            if word in key:
                pos = convert_dict[key]
        return pos
            # else:
            #     pos = ''
    if parser == 'mystem':
        for w in word:
            if 'analysis' in w:
                gr = w['analysis'][0]['gr']
                pos = gr.split('=')[0].split(',')[0]
            else:
                pos = ''
        for key in convert_dict:
            if pos in key:
                pos = convert_dict[key]
    return pos
```

```python
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()
```

```python
pymorphy_output = []
```

```python
for word in golden_standard_list:
    word_pair = []
    word = word[0]
    word_analyzed = morph.parse(word)
    word_analyzed_standardized = convert_to_standard(
        word_analyzed[0], parser='pymorphy')
    
    word_pair = [word, word_analyzed_standardized]
    pymorphy_output.append(word_pair)
```

```python
pymorphy_output
```

```python
from pymystem3 import Mystem
mystem_morph = Mystem()
```

```python
mystem_output = []
```

```python
for word in golden_standard_list:
    word_pair = []
    word = word[0]
    word_analyzed = mystem_morph.analyze(word)
    word_analyzed_standardized = convert_to_standard(
        word_analyzed, parser='mystem')

    word_pair = [word, word_analyzed_standardized]
    mystem_output.append(word_pair)
```

```python
import spacy
from spacy.lang.ru.examples import sentences

nlp = spacy.load("ru_core_news_sm")
```

```python
spacy_output = []
```

```python
for word in golden_standard_list:
    word_pair = []
    word = word[0]
    word_analyzed = nlp(word)
    word_pos = word_analyzed[0].pos_
    word_analyzed_standardized = convert_to_standard(
        word_pos, parser='spacy')

    word_pair = [word, word_analyzed_standardized]
    spacy_output.append(word_pair)
```

```python
spacy_output
```

```python
def accuracy_counter(my_output, standard):
    is_same = 0
    for i in range(len(standard)):
        if my_output[i][1] == standard[i][1]:
            is_same += 1
    return is_same / len(standard)
```

```python
accuracy_counter(pymorphy_output, golden_standard_list)
```

```python
accuracy_counter(mystem_output, golden_standard_list)
```

```python
accuracy_counter(spacy_output, golden_standard_list)
```
